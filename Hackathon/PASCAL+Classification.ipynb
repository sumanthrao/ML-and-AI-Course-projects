```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
from keras.callbacks import EarlyStopping
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.utils import np_utils
```

```python
temp = []
for i in labels:
    temp2 = {}
    temp2['Image'] = i
    temp2['Y'] = np.array(labels[i]).astype("float32")
    temp.append(temp2)
temp[0:3]
```

```python
output_df = pd.DataFrame(temp)
```

```python
output_df.head()
```

```python
joined_df = pd.merge(train, output_df, on='Image')
```

```python
joined_df.loc[[0]]
```

```python
new_train_X = joined_df.iloc[:,1:-1]
new_train_Y = joined_df.iloc[:,-1]
```

```python
new_train_Y.head()
```

```python
seed = 123 
X_train, X_test, y_train, y_test = train_test_split(new_train_X, new_train_Y, test_size=0.20, random_state=seed)

X_train = X_train/255
X_test = X_test/255
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_dash = X_train.values.reshape(X_train.shape[0], 100, 100, 1).astype('float32')
X_test_dash = X_test.values.reshape(X_test.shape[0], 100, 100, 1).astype('float32')
```

```python
X_dash.shape
```

```python
np.array(y_train)
```

```python
y_train[0].shape
```

```python
l = []
for x in y_train:
    l.append(list(x))
Y = np.array(l)
```

```python
Y.shape
```

```python
Y
```

```python
def cnn_model():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), input_shape=(100, 100, 1), 
                     activation='relu', name="conv_1")) # 32 implies 32 bit image
    model.add(MaxPooling2D(pool_size=(2, 2), name="pool_1"))
    model.add(Conv2D(64, (4, 4), activation='relu', name="conv_2"))
    model.add(MaxPooling2D(pool_size=(4, 4), name="pool_2"))
    model.add(Dropout(0.3))
    model.add(Flatten())
    model.add(Dense(150, activation='relu', name="dense_1"))
    model.add(Dense(80, activation='relu', name="dense_2"))
    model.add(Dense(20, activation='sigmoid', name="modeloutput"))
    
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model = cnn_model()

early_stopping_monitor = EarlyStopping(patience=2)

epochs = 15
batch_size = 200
history = model.fit(X_dash, Y, validation_split=0.20, epochs=epochs, batch_size=batch_size, 
                    callbacks=[early_stopping_monitor], verbose=2)
```

```python
X_test_dash.shape
```

```python
m = []
for x in y_test:
    m.append(list(x))
Y_test = np.array(m)
```

```python
Y_test.shape
```

```python
scores = model.evaluate(X_test_dash, Y_test, verbose=2)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
